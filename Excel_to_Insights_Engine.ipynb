```python
# ===== EXPORT INSIGHTS TO TEXT FILE =====

insight_path = "/content/drive/MyDrive/GColab/final_insights_summary.txt"

with open(insight_path, "w") as f:
    f.write("===== FINAL INSIGHTS SUMMARY =====\n\n")

    f.write(f"Total cleaned rows: {len(combined_df)}\n")
    f.write(f"Columns: {list(combined_df.columns)}\n\n")

    if "Amount" in combined_df.columns:
        f.write(f"Total Amount: {combined_df['Amount'].sum()}\n")
        f.write(f"Average Amount: {combined_df['Amount'].mean():.2f}\n")
        f.write(f"Highest Amount: {combined_df['Amount'].max()}\n")
        f.write(f"Lowest Amount: {combined_df['Amount'].min()}\n\n")

    if "Category" in combined_df.columns and "Amount" in combined_df.columns:
        category_summary = combined_df.groupby("Category")["Amount"].sum().sort_values(ascending=False)

        top_category = category_summary.idxmax()
        top_value = category_summary.max()

        low_category = category_summary.idxmin()
        low_value = category_summary.min()

        f.write("Category-wise Summary:\n")
        f.write(str(category_summary))
        f.write("\n\n")

        f.write(f"Top Category: {top_category} ({top_value})\n")
        f.write(f"Lowest Category: {low_category} ({low_value})\n\n")

    if "Source_File" in combined_df.columns and "Amount" in combined_df.columns:
        source_summary = combined_df.groupby("Source_File")["Amount"].sum().sort_values(ascending=False)

        top_source = source_summary.idxmax()
        top_source_value = source_summary.max()

        f.write("Source-wise Summary:\n")
        f.write(str(source_summary))
        f.write("\n\n")

        f.write(f"Top Source File: {top_source} ({top_source_value})\n\n")

    f.write("===== HUMAN STYLE SUMMARY =====\n")
    if "Category" in combined_df.columns and "Amount" in combined_df.columns:
        f.write(f"The dataset contains {len(combined_df)} cleaned records.\n")
        f.write(f"The total amount across all records is {combined_df['Amount'].sum()}.\n")
        f.write(f"The average amount per record is {combined_df['Amount'].mean():.2f}.\n")
        f.write(f"The strongest category is {top_category}, contributing {top_value}.\n")
        f.write(f"The weakest category is {low_category}, contributing {low_value}.\n")
        f.write("This indicates where the data is currently strongest and which category may need more attention.\n")

print("✅ Insights saved to:", insight_path)
```

```python
# ===== EXPORT FINAL REPORT WITH INSIGHTS SHEET =====

output_path = "/content/drive/MyDrive/GColab/final_multi_file_report_with_insights.xlsx"

insights_data = []

# Basic stats
insights_data.append(["Metric", "Value"])
insights_data.append(["Total Cleaned Rows", len(combined_df)])
insights_data.append(["Columns", ", ".join(combined_df.columns)])

if "Amount" in combined_df.columns:
    insights_data.append(["Total Amount", combined_df["Amount"].sum()])
    insights_data.append(["Average Amount", round(combined_df["Amount"].mean(), 2)])
    insights_data.append(["Highest Amount", combined_df["Amount"].max()])
    insights_data.append(["Lowest Amount", combined_df["Amount"].min()])

if "Category" in combined_df.columns and "Amount" in combined_df.columns:
    category_summary = combined_df.groupby("Category")["Amount"].sum().sort_values(ascending=False)

    top_category = category_summary.idxmax()
    top_value = category_summary.max()

    low_category = category_summary.idxmin()
    low_value = category_summary.min()

    insights_data.append(["Top Category", top_category])
    insights_data.append(["Top Category Total", top_value])
    insights_data.append(["Lowest Category", low_category])
    insights_data.append(["Lowest Category Total", low_value])

if "Source_File" in combined_df.columns and "Amount" in combined_df.columns:
    source_summary = combined_df.groupby("Source_File")["Amount"].sum().sort_values(ascending=False)

    top_source = source_summary.idxmax()
    top_source_value = source_summary.max()

    insights_data.append(["Top Source File", top_source])
    insights_data.append(["Top Source Total", top_source_value])

# Human summary lines
human_summary = []
human_summary.append("Human Style Summary")

if "Category" in combined_df.columns and "Amount" in combined_df.columns:
    human_summary.append(f"The dataset contains {len(combined_df)} cleaned records.")
    human_summary.append(f"The total amount across all records is {combined_df['Amount'].sum()}.")
    human_summary.append(f"The average amount per record is {combined_df['Amount'].mean():.2f}.")
    human_summary.append(f"The strongest category is {top_category}, contributing {top_value}.")
    human_summary.append(f"The weakest category is {low_category}, contributing {low_value}.")
    human_summary.append("This indicates where the data is currently strongest and which category may need more attention.")

insights_df = pd.DataFrame(insights_data[1:], columns=insights_data[0])
human_summary_df = pd.DataFrame({"Summary": human_summary})

with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    combined_df.to_excel(writer, sheet_name="Combined_Data", index=False)

    if "Source_File" in combined_df.columns:
        source_count_summary = combined_df["Source_File"].value_counts().reset_index()
        source_count_summary.columns = ["Source_File", "Row_Count"]
        source_count_summary.to_excel(writer, sheet_name="Source_Summary", index=False)

    if "Category" in combined_df.columns and "Amount" in combined_df.columns:
        category_summary.reset_index().to_excel(writer, sheet_name="Category_Summary", index=False)

    insights_df.to_excel(writer, sheet_name="Key_Insights", index=False)
    human_summary_df.to_excel(writer, sheet_name="Human_Summary", index=False)

print("✅ Final report with insights saved to:", output_path)
```

```python
# ===== EXPORT FORMATTED FINAL REPORT =====

import pandas as pd
from openpyxl.styles import Font, PatternFill, Alignment
from openpyxl.utils import get_column_letter

output_path = "/content/drive/MyDrive/GColab/final_multi_file_report_formatted.xlsx"

# Prepare insight tables
insights_data = []

insights_data.append(["Metric", "Value"])
insights_data.append(["Total Cleaned Rows", len(combined_df)])
insights_data.append(["Columns", ", ".join(combined_df.columns)])

if "Amount" in combined_df.columns:
    insights_data.append(["Total Amount", combined_df["Amount"].sum()])
    insights_data.append(["Average Amount", round(combined_df["Amount"].mean(), 2)])
    insights_data.append(["Highest Amount", combined_df["Amount"].max()])
    insights_data.append(["Lowest Amount", combined_df["Amount"].min()])

if "Category" in combined_df.columns and "Amount" in combined_df.columns:
    category_summary = combined_df.groupby("Category")["Amount"].sum().sort_values(ascending=False)
    top_category = category_summary.idxmax()
    top_value = category_summary.max()
    low_category = category_summary.idxmin()
    low_value = category_summary.min()

    insights_data.append(["Top Category", top_category])
    insights_data.append(["Top Category Total", top_value])
    insights_data.append(["Lowest Category", low_category])
    insights_data.append(["Lowest Category Total", low_value])

if "Source_File" in combined_df.columns and "Amount" in combined_df.columns:
    source_summary = combined_df.groupby("Source_File")["Amount"].sum().sort_values(ascending=False)
    top_source = source_summary.idxmax()
    top_source_value = source_summary.max()

    insights_data.append(["Top Source File", top_source])
    insights_data.append(["Top Source Total", top_source_value])

human_summary = []
human_summary.append("Human Style Summary")

if "Category" in combined_df.columns and "Amount" in combined_df.columns:
    human_summary.append(f"The dataset contains {len(combined_df)} cleaned records.")
    human_summary.append(f"The total amount across all records is {combined_df['Amount'].sum()}.")
    human_summary.append(f"The average amount per record is {combined_df['Amount'].mean():.2f}.")
    human_summary.append(f"The strongest category is {top_category}, contributing {top_value}.")
    human_summary.append(f"The weakest category is {low_category}, contributing {low_value}.")
    human_summary.append("This indicates where the data is currently strongest and which category may need more attention.")

insights_df = pd.DataFrame(insights_data[1:], columns=insights_data[0])
human_summary_df = pd.DataFrame({"Summary": human_summary})

# Write sheets
with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    combined_df.to_excel(writer, sheet_name="Combined_Data", index=False)

    if "Source_File" in combined_df.columns:
        source_count_summary = combined_df["Source_File"].value_counts().reset_index()
        source_count_summary.columns = ["Source_File", "Row_Count"]
        source_count_summary.to_excel(writer, sheet_name="Source_Summary", index=False)

    if "Category" in combined_df.columns and "Amount" in combined_df.columns:
        category_summary.reset_index().to_excel(writer, sheet_name="Category_Summary", index=False)

    insights_df.to_excel(writer, sheet_name="Key_Insights", index=False)
    human_summary_df.to_excel(writer, sheet_name="Human_Summary", index=False)

    # Formatting
    wb = writer.book
    header_fill = PatternFill(fill_type="solid", fgColor="D9EAF7")
    header_font = Font(bold=True)
    center_align = Alignment(vertical="center")

    for sheet_name in wb.sheetnames:
        ws = wb[sheet_name]

        # Freeze top row
        ws.freeze_panes = "A2"

        # Format header row
        for cell in ws[1]:
            cell.font = header_font
            cell.fill = header_fill
            cell.alignment = center_align

        # Auto-fit column widths
        for col_cells in ws.columns:
            max_length = 0
            col_letter = get_column_letter(col_cells[0].column)

            for cell in col_cells:
                try:
                    cell_len = len(str(cell.value)) if cell.value is not None else 0
                    if cell_len > max_length:
                        max_length = cell_len
                except:
                    pass

            ws.column_dimensions[col_letter].width = min(max_length + 2, 40)

print("✅ Formatted final report saved to:", output_path)
```
